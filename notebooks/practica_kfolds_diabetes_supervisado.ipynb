## 1. Marco teórico y analítico

### 1.1 Aprendizaje supervisado y clasificación binaria

El aprendizaje supervisado permite construir modelos predictivos a partir de datos previamente etiquetados. En un problema de clasificación, se dispone de una base de datos:

$$
D = \{t_1, t_2, ..., t_n\}
$$

formada por registros o individuos, y de un conjunto de clases:

$$
C = \{C_1, C_2, ..., C_m\}
$$

El objetivo consiste en encontrar una función:

$$
f:D \rightarrow C
$$

capaz de asignar cada registro $t_i$ a una clase $C_j$. En esta práctica, la función de clasificación se aproxima mediante tres máquinas de aprendizaje supervisado: **Random Forest**, **Máquina de Soporte Vectorial** y **Red Neuronal Multicapa**. La comparación se realiza con **K-Fold Cross-Validation (K-Folds)** para estimar el comportamiento predictivo de los modelos mediante particiones reproducibles y estratificadas.


### 1.2 Dataset de diabetes

Se utiliza el archivo `diabetes.csv`, compuesto por 768 registros y nueve columnas. Cada registro representa una observación clínica descrita mediante variables numéricas relacionadas con embarazos, glucosa, presión arterial, grosor de piel, insulina, índice de masa corporal, función de pedigrí diabético, edad y una variable objetivo binaria.

La variable objetivo es `Outcome`, donde:

- `0`: paciente sin diagnóstico positivo de diabetes.
- `1`: paciente con diagnóstico positivo de diabetes.

El uso de este dataset permite plantear un problema supervisado de clasificación binaria, en el que los modelos deben aprender patrones clínicos asociados con la presencia o ausencia de diabetes. Esta práctica tiene finalidad académica y no debe interpretarse como sistema clínico de diagnóstico.


### 1.3 K-Fold Cross-Validation (K-Folds)

K-Fold Cross-Validation es una técnica de validación cruzada que divide el conjunto de datos en $k$ subconjuntos o *folds* de tamaño similar. En cada iteración, uno de los folds se utiliza como conjunto de prueba y los $k-1$ folds restantes se emplean para entrenar el modelo. El proceso se repite $k$ veces, de manera que cada fold funciona exactamente una vez como conjunto de prueba.

El error promedio de K-Folds puede representarse como:

$$
CV_k = \frac{1}{k}\sum_{i=1}^{k}E_i
$$

donde $E_i$ es el error obtenido en el fold $i$.

En esta práctica se utiliza **K-Folds estratificado con $k=10$**, una variante adecuada para clasificación binaria porque conserva aproximadamente la proporción de clases en cada partición. Esta decisión permite comparar Random Forest, SVM Lineal y Red Neuronal MLP bajo una estrategia reproducible, equilibrada y computacionalmente eficiente y robusta para clasificación supervisada.

### 1.4 Random Forest

Random Forest es un método de ensamble que combina múltiples árboles de decisión. Cada árbol se entrena sobre una muestra bootstrap del conjunto de datos y, en cada división, considera un subconjunto aleatorio de variables. La predicción final se obtiene por votación mayoritaria en clasificación.

En esta práctica, Random Forest es pertinente porque puede capturar relaciones no lineales entre las variables clínicas y la clase `Outcome`. Además, reduce la varianza de un árbol individual mediante agregación, lo que lo convierte en un modelo robusto para clasificación tabular.


### 1.5 Máquina de Soporte Vectorial (SVM)

La Máquina de Soporte Vectorial es un clasificador supervisado que busca construir un hiperplano de separación entre clases maximizando el margen geométrico. Los puntos más cercanos al hiperplano se denominan vectores de soporte, porque determinan la frontera de decisión.

En esta práctica se utiliza una SVM Lineal dentro de un pipeline con estandarización. La estandarización es necesaria porque SVM depende de escalas numéricas y distancias geométricas en el espacio de características.


### 1.6 Red Neuronal Multicapa (MLP)

Una red neuronal multicapa es un modelo supervisado no lineal compuesto por capas de neuronas artificiales. Cada neurona aplica una combinación ponderada de entradas y una función de activación. Durante el entrenamiento, el modelo ajusta sus pesos para reducir el error de clasificación.

En esta práctica se implementa una arquitectura compacta mediante `MLPClassifier`, adecuada para datos tabulares de tamaño moderado. Al igual que SVM, requiere estandarización de variables para mejorar la estabilidad del entrenamiento.


### 1.7 Métricas solicitadas para la prueba de modelos

La evaluación comparativa se realiza a partir de la matriz de confusión:

| Elemento | Significado |
|---|---|
| TP | Verdaderos positivos: pacientes positivos correctamente clasificados. |
| TN | Verdaderos negativos: pacientes negativos correctamente clasificados. |
| FP | Falsos positivos: pacientes negativos clasificados como positivos. |
| FN | Falsos negativos: pacientes positivos clasificados como negativos. |

A partir de estos elementos se calculan:

$$
Precisión\ Global = \frac{TP + TN}{TP + TN + FP + FN}
$$

$$
Error\ Global = \frac{FP + FN}{TP + TN + FP + FN}
$$

$$
PP = \frac{TP}{TP + FN}
$$

$$
PN = \frac{TN}{TN + FP}
$$

$$
AP = \frac{TP}{TP + FP}
$$

$$
NP = \frac{TN}{TN + FN}
$$

donde **PP** representa la precisión positiva o sensibilidad, **PN** representa la precisión negativa o especificidad, **AP** representa la asertividad positiva y **NP** representa la asertividad negativa.

```python
# Funciones de carga, validación y visualización HTML del dataset.

REQUIRED_COLUMNS = [
    "Pregnancies",
    "Glucose",
    "BloodPressure",
    "SkinThickness",
    "Insulin",
    "BMI",
    "DiabetesPedigreeFunction",
    "Age",
    "Outcome",
]

FEATURE_COLUMNS = [
    "Pregnancies",
    "Glucose",
    "BloodPressure",
    "SkinThickness",
    "Insulin",
    "BMI",
    "DiabetesPedigreeFunction",
    "Age",
]


def load_diabetes_dataset(csv_path: Path) -> pd.DataFrame:
    """Carga el dataset de diabetes desde un archivo CSV.

    Esta función representa la fase de ingestión de datos del pipeline.
    Lee el archivo original, valida que exista y devuelve un DataFrame
    para las etapas posteriores de validación, exploración y modelado.

    Args:
        csv_path: Ruta del archivo CSV `diabetes.csv`.

    Returns:
        DataFrame con los registros clínicos cargados.

    Raises:
        FileNotFoundError: Si el archivo CSV no existe.
        ValueError: Si el archivo existe pero no contiene registros.
        RuntimeError: Si ocurre un error inesperado durante la lectura.
    """
    if not isinstance(csv_path, Path):
        raise TypeError("csv_path debe ser una instancia de pathlib.Path.")

    if not csv_path.exists():
        raise FileNotFoundError(f"No se encontró el archivo: {csv_path}")

    try:
        df_loaded = pd.read_csv(csv_path)
    except Exception as exc:
        LOGGER.exception("Error al cargar el dataset.")
        raise RuntimeError("No fue posible cargar el dataset de diabetes.") from exc

    if df_loaded.empty:
        raise ValueError("El dataset está vacío.")

    LOGGER.info("Dataset cargado correctamente desde %s", format_relative_path(csv_path))
    return df_loaded


def validate_diabetes_dataset(df_input: pd.DataFrame, required_columns: Iterable[str]) -> None:
    """Valida la estructura mínima del dataset de diabetes.

    La validación revisa columnas obligatorias, valores faltantes,
    variable objetivo binaria y tipos numéricos. Esta etapa evita entrenar
    modelos sobre datos inconsistentes.

    Args:
        df_input: DataFrame cargado desde el archivo CSV.
        required_columns: Colección de nombres de columnas obligatorias.

    Returns:
        None. La función finaliza sin retorno si la validación es correcta.

    Raises:
        TypeError: Si df_input no es un DataFrame.
        ValueError: Si faltan columnas, hay valores faltantes, la variable
        objetivo no es binaria o existen columnas no numéricas.
    """
    if not isinstance(df_input, pd.DataFrame):
        raise TypeError("df_input debe ser un pandas DataFrame.")

    missing_columns = [col for col in required_columns if col not in df_input.columns]
    if missing_columns:
        raise ValueError(f"Faltan columnas obligatorias: {missing_columns}")

    if df_input[list(required_columns)].isna().sum().sum() > 0:
        raise ValueError("El dataset contiene valores faltantes en columnas obligatorias.")

    outcome_values = sorted(df_input["Outcome"].unique().tolist())
    if outcome_values != [0, 1]:
        raise ValueError(f"La variable Outcome debe ser binaria con valores [0, 1]. Valores detectados: {outcome_values}")

    non_numeric_columns = [
        col for col in required_columns
        if not pd.api.types.is_numeric_dtype(df_input[col])
    ]
    if non_numeric_columns:
        raise ValueError(f"Todas las columnas deben ser numéricas. Columnas no numéricas: {non_numeric_columns}")

    LOGGER.info("Validación estructural completada correctamente.")


def display_scrollable_dataframe(
    df_input: pd.DataFrame,
    height_px: int = 520,
    title: str = "Dataset completo",
) -> None:
    """Muestra un DataFrame en una ventana HTML responsiva con índice y scroll.

    Esta visualización replica el estilo operativo de la práctica base:
    el dataset se muestra de forma transparente dentro del notebook sin
    saturar la página completa. Además, agrega una columna `registro`
    para identificar la secuencia completa de observaciones cargadas.

    Args:
        df_input: DataFrame que se desea visualizar.
        height_px: Altura de la ventana de scroll en pixeles.
        title: Título descriptivo mostrado antes de la tabla.

    Returns:
        None. Renderiza una salida HTML dentro del notebook.

    Raises:
        TypeError: Si df_input no es un DataFrame.
        ValueError: Si height_px no es positivo.
    """
    if not isinstance(df_input, pd.DataFrame):
        raise TypeError("df_input debe ser un pandas DataFrame.")

    if height_px <= 0:
        raise ValueError("height_px debe ser un entero positivo.")

    df_display = df_input.copy()
    if "registro" not in df_display.columns:
        df_display.insert(0, "registro", range(1, len(df_display) + 1))

    html_table = df_display.to_html(index=False)

    html_scroll = f"""
    <h4 style="margin-top: 12px; margin-bottom: 10px;">{title}</h4>
    <div class="responsive-table-wrapper" style="
        max-height: {height_px}px;
        max-width: 100%;
        overflow-x: auto;
        overflow-y: auto;
        border: 1px solid #b8c2cc;
        border-radius: 6px;
        font-family: Arial, sans-serif;
        -webkit-overflow-scrolling: touch;
    ">
    <style>
        .responsive-table-wrapper table {{
            border-collapse: collapse;
            width: 100%;
            min-width: 760px;
            font-size: clamp(11px, 1.6vw, 14px);
            color: #111111 !important;
            background-color: #ffffff !important;
        }}
        .responsive-table-wrapper th {{
            position: sticky;
            top: 0;
            z-index: 2;
            background-color: #e8eef5 !important;
            color: #000000 !important;
            font-weight: 700;
            text-align: center;
            padding: 10px;
            border-bottom: 1px solid #b8c2cc;
            white-space: nowrap;
        }}
        .responsive-table-wrapper td {{
            padding: 9px;
            text-align: center;
            border-bottom: 1px solid #d9dee5;
            color: #111111 !important;
            font-weight: 500;
            white-space: nowrap;
        }}
        .responsive-table-wrapper td:first-child {{
            color: #000000 !important;
            font-weight: 700;
            background-color: #eef4fb !important;
        }}
        .responsive-table-wrapper tr:nth-child(even) td {{
            background-color: #f3f6fa !important;
        }}
        .responsive-table-wrapper tr:nth-child(odd) td {{
            background-color: #ffffff !important;
        }}
        .responsive-table-wrapper tr:hover td {{
            background-color: #dcecff !important;
            color: #000000 !important;
        }}

        @media (max-width: 768px) {{
            .responsive-table-wrapper {{
                max-height: 420px !important;
            }}
            .responsive-table-wrapper table {{
                min-width: 720px;
            }}
            .responsive-table-wrapper th,
            .responsive-table-wrapper td {{
                padding: 7px;
            }}
        }}
    </style>
    {html_table}
    </div>
    """

    display(HTML(html_scroll))



def display_responsive_table(
    df_input: pd.DataFrame,
    title: str = "Tabla",
    min_width_px: int = 720,
    index: bool = False,
) -> None:
    """Muestra una tabla HTML con desplazamiento horizontal responsivo.

    Esta función evita que las tablas anchas se desborden en el notebook
    o en el HTML exportado. El contenido queda dentro de un contenedor
    con scroll horizontal y conserva un estilo visual coherente con la
    práctica.

    Args:
        df_input: DataFrame que se desea mostrar.
        title: Título descriptivo de la tabla.
        min_width_px: Ancho mínimo de la tabla para activar scroll horizontal.
        index: Indica si se conserva el índice del DataFrame.

    Returns:
        None. Renderiza una tabla HTML responsiva.

    Raises:
        TypeError: Si df_input no es un DataFrame.
        ValueError: Si min_width_px no es positivo.
    """
    if not isinstance(df_input, pd.DataFrame):
        raise TypeError("df_input debe ser un pandas DataFrame.")

    if min_width_px <= 0:
        raise ValueError("min_width_px debe ser un entero positivo.")

    html_table = df_input.to_html(index=index)

    html = f"""
    <h4>{title}</h4>
    <div class="responsive-table-wrapper" style="
        width: 100%;
        max-width: 100%;
        overflow-x: auto;
        overflow-y: hidden;
        border: 1px solid #b8c2cc;
        border-radius: 6px;
        margin: 12px 0 16px 0;
        -webkit-overflow-scrolling: touch;
    ">
    <style>
        .responsive-table-wrapper table {{
            border-collapse: collapse;
            width: 100%;
            min-width: {min_width_px}px;
            font-size: 14px;
            color: #111111 !important;
            background-color: #ffffff !important;
        }}
        .responsive-table-wrapper th {{
            background-color: #e8eef5 !important;
            color: #000000 !important;
            font-weight: 700;
            text-align: center;
            padding: 9px;
            border: 1px solid #b8c2cc;
            white-space: nowrap;
        }}
        .responsive-table-wrapper td {{
            padding: 9px;
            text-align: center;
            border: 1px solid #d9dee5;
            color: #111111 !important;
            font-weight: 500;
            white-space: nowrap;
        }}
        .responsive-table-wrapper tr:nth-child(even) td {{
            background-color: #f3f6fa !important;
        }}
        .responsive-table-wrapper tr:nth-child(odd) td {{
            background-color: #ffffff !important;
        }}
        .responsive-table-wrapper tr:hover td {{
            background-color: #dcecff !important;
            color: #000000 !important;
        }}
        @media (max-width: 768px) {{
            .responsive-table-wrapper table {{
                min-width: {min_width_px}px;
                font-size: 13px;
            }}
            .responsive-table-wrapper th,
            .responsive-table-wrapper td {{
                padding: 7px;
            }}
        }}
    </style>
    {html_table}
    </div>
    """
    display(HTML(html))


def display_responsive_image(image_path: Path, max_width_px: int = 650) -> None:
    """Muestra una imagen PNG de forma responsiva dentro del notebook y HTML exportado.

    La función incrusta la imagen en base64 para que el reporte HTML sea
    portable en GitHub Pages y ajusta su tamaño con CSS para que no exceda
    el ancho disponible de pantalla.

    Args:
        image_path: Ruta de la imagen PNG generada por el pipeline.
        max_width_px: Ancho máximo recomendado en pixeles.

    Returns:
        None. Renderiza la imagen como HTML responsivo.

    Raises:
        TypeError: Si image_path no es una instancia de pathlib.Path.
        FileNotFoundError: Si la imagen no existe.
        ValueError: Si max_width_px no es positivo.
    """
    if not isinstance(image_path, Path):
        raise TypeError("image_path debe ser una instancia de pathlib.Path.")

    if not image_path.exists():
        raise FileNotFoundError(f"No se encontró la imagen: {image_path}")

    if max_width_px <= 0:
        raise ValueError("max_width_px debe ser un entero positivo.")

    encoded_image = base64.b64encode(image_path.read_bytes()).decode("utf-8")

    image_html = f"""
    <div style="width: 100%; text-align: center; margin: 12px 0;">
        <img
            src="data:image/png;base64,{encoded_image}"
            alt="{image_path.name}"
            style="
                width: min(100%, {max_width_px}px);
                max-width: 100%;
                height: auto;
                display: inline-block;
            "
        />
    </div>
    """
    display(HTML(image_html))
```

Para evitar fuga de datos, el escalamiento de SVM y MLP se integra dentro de `Pipeline`, de modo que el `StandardScaler` se ajusta únicamente con los datos de entrenamiento de cada fold.

## 8. Implementación de las máquinas de aprendizaje supervisado con K-Folds

Se implementan tres modelos predictivos:

1. **Random Forest**: ensamble de árboles con votación mayoritaria.
2. **SVM Lineal**: clasificador de margen máximo con estandarización.
3. **Red Neuronal MLP**: modelo no lineal con una capa oculta compacta y estandarización.

Todos los modelos se evalúan con el mismo objeto `StratifiedKFold`, garantizando que la comparación sea homogénea y que la proporción de clases se mantenga aproximadamente en cada fold.

```python
# Funciones de construcción de modelos y evaluación K-Folds.

def build_model_pipelines(random_state: int = RANDOM_STATE) -> dict[str, object]:
    """Construye los modelos supervisados utilizados en la práctica.

    La función separa la configuración de algoritmos del proceso de evaluación.
    SVM y MLP se implementan como pipelines con StandardScaler para evitar
    fuga de datos durante K-Folds. Random Forest se usa sin escalamiento porque
    los árboles de decisión no dependen de la escala de las variables.

    Args:
        random_state: Semilla para reproducibilidad experimental.

    Returns:
        Diccionario donde cada clave es el nombre del modelo y cada valor es
        un estimador compatible con scikit-learn.

    Raises:
        ValueError: Si random_state no es entero no negativo.
    """
    if not isinstance(random_state, int) or random_state < 0:
        raise ValueError("random_state debe ser un entero no negativo.")

    models = {
        "Random Forest": RandomForestClassifier(
            n_estimators=15,
            max_depth=5,
            random_state=random_state,
            n_jobs=1,
            class_weight="balanced",
        ),
        "SVM Lineal": Pipeline(
            steps=[
                ("scaler", StandardScaler()),
                ("model", LinearSVC(
                    C=1.0,
                    random_state=random_state,
                    max_iter=2000,
                    dual=False,
                )),
            ]
        ),
        "Red Neuronal MLP": Pipeline(
            steps=[
                ("scaler", StandardScaler()),
                ("model", MLPClassifier(
                    hidden_layer_sizes=(8,),
                    activation="relu",
                    solver="adam",
                    alpha=0.001,
                    max_iter=40,
                    random_state=random_state,
                    early_stopping=False,
                )),
            ]
        ),
    }

    LOGGER.info("Modelos construidos: %s", list(models.keys()))
    return models


def run_kfolds_predictions(
    models: dict[str, object],
    X_input: pd.DataFrame,
    y_input: pd.Series,
    n_splits: int = N_SPLITS,
    random_state: int = RANDOM_STATE,
) -> dict[str, np.ndarray]:
    """Ejecuta K-Folds estratificado para cada modelo y obtiene predicciones fuera de muestra.

    Para cada algoritmo, la función divide el dataset en k folds. En cada
    iteración, un fold queda como prueba y los folds restantes se usan para
    entrenamiento. El resultado es un vector de predicciones comparable con y_input.

    Args:
        models: Diccionario de estimadores supervisados.
        X_input: Matriz de variables predictoras.
        y_input: Vector de clases reales.
        n_splits: Número de folds usados en la validación cruzada.
        random_state: Semilla para reproducibilidad.

    Returns:
        Diccionario con las predicciones K-Folds de cada modelo.

    Raises:
        TypeError: Si models no es un diccionario.
        ValueError: Si X_input y y_input tienen tamaños incompatibles o n_splits es inválido.
        RuntimeError: Si ocurre un error durante la validación cruzada.
    """
    if not isinstance(models, dict) or not models:
        raise TypeError("models debe ser un diccionario no vacío.")

    if len(X_input) != len(y_input):
        raise ValueError("X_input y y_input deben tener el mismo número de registros.")

    if not isinstance(n_splits, int) or n_splits < 2:
        raise ValueError("n_splits debe ser un entero mayor o igual a 2.")

    kfolds = StratifiedKFold(
        n_splits=n_splits,
        shuffle=True,
        random_state=random_state,
    )
    predictions = {}

    for model_name, estimator in models.items():
        try:
            LOGGER.info("Iniciando K-Folds para: %s", model_name)
            y_pred = cross_val_predict(
                estimator=estimator,
                X=X_input,
                y=y_input,
                cv=kfolds,
                n_jobs=1,
                method="predict",
            )
            predictions[model_name] = y_pred
            LOGGER.info("K-Folds finalizado para: %s", model_name)
        except Exception as exc:
            LOGGER.exception("Error durante K-Folds en el modelo %s", model_name)
            raise RuntimeError(f"No fue posible ejecutar K-Folds para {model_name}.") from exc

    return predictions
```

### Aclaración sobre el entrenamiento en K-Folds

En K-Folds cada modelo se entrena tantas veces como folds se definan. En esta práctica se utiliza `k = 10`, por lo que cada algoritmo se entrena 10 veces: en cada iteración utiliza aproximadamente el 90% de los registros para entrenamiento y el 10% para prueba.

Por esta razón, aunque no se realiza una única partición fija entrenamiento/prueba, sí existe entrenamiento supervisado en cada iteración del procedimiento K-Folds.

```python
# Ejecución de K-Folds.

models = build_model_pipelines(random_state=RANDOM_STATE)
cache_path = OUTPUT_TABLES / "predicciones_kfolds_cache.csv"

if cache_path.exists():
    LOGGER.info("Se encontró caché de predicciones K-Folds: %s", format_relative_path(cache_path))
    prediction_cache_df = pd.read_csv(cache_path)
    kfolds_predictions = {
        "Random Forest": prediction_cache_df["Random Forest"].to_numpy(),
        "SVM Lineal": prediction_cache_df["SVM Lineal"].to_numpy(),
        "Red Neuronal MLP": prediction_cache_df["Red Neuronal MLP"].to_numpy(),
    }
else:
    LOGGER.info("No se encontró caché. Se ejecutará K-Folds completo.")
    kfolds_predictions = run_kfolds_predictions(models, X, y, n_splits=N_SPLITS, random_state=RANDOM_STATE)
    prediction_cache_df = pd.DataFrame({"Outcome_real": y})
    for model_name, y_pred in kfolds_predictions.items():
        prediction_cache_df[model_name] = y_pred
    prediction_cache_df.to_csv(cache_path, index=False)

print("Predicciones K-Folds disponibles correctamente.")
for model_name, y_pred in kfolds_predictions.items():
    print(f"{model_name}: {len(y_pred)} predicciones")

print(f"Caché K-Folds: {format_relative_path(cache_path)}")
```

## 9. Prueba de las máquinas de aprendizaje supervisado

La prueba se realiza mediante las predicciones obtenidas con **K-Folds**. Para cada modelo se construye una matriz de confusión y se calculan las métricas solicitadas:

- Precisión Global.
- Error Global.
- Precisión Positiva (PP).
- Precisión Negativa (PN).
- Falsos Positivos (FP).
- Falsos Negativos (FN).
- Asertividad Positiva (AP).
- Asertividad Negativa (NP).

```python
# Funciones de cálculo de métricas solicitadas.

def safe_divide(numerator: float, denominator: float) -> float:
    """Divide dos valores evitando errores por división entre cero.

    Args:
        numerator: Numerador de la operación.
        denominator: Denominador de la operación.

    Returns:
        Resultado de la división. Si el denominador es cero, devuelve 0.0.
    """
    if denominator == 0:
        return 0.0
    return numerator / denominator


def calculate_requested_metrics(
    y_true: pd.Series,
    y_pred: np.ndarray,
    model_name: str,
) -> dict[str, float | int | str]:
    """Calcula métricas de clasificación solicitadas por la práctica.

    La matriz de confusión se interpreta con la clase 0 como negativa y
    la clase 1 como positiva. A partir de TN, FP, FN y TP se calculan
    precisión global, error global, precisión positiva, precisión negativa,
    asertividad positiva y asertividad negativa.

    Args:
        y_true: Etiquetas reales.
        y_pred: Etiquetas predichas mediante K-Folds.
        model_name: Nombre descriptivo del modelo evaluado.

    Returns:
        Diccionario con métricas y conteos principales.

    Raises:
        ValueError: Si los vectores de entrada tienen longitudes distintas.
    """
    if len(y_true) != len(y_pred):
        raise ValueError("y_true y y_pred deben tener la misma longitud.")

    cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
    tn, fp, fn, tp = cm.ravel()
    total = tn + fp + fn + tp

    metrics = {
        "Modelo": model_name,
        "TN": int(tn),
        "FP": int(fp),
        "FN": int(fn),
        "TP": int(tp),
        "Precisión Global": safe_divide(tp + tn, total),
        "Error Global": safe_divide(fp + fn, total),
        "PP": safe_divide(tp, tp + fn),
        "PN": safe_divide(tn, tn + fp),
        "AP": safe_divide(tp, tp + fp),
        "NP": safe_divide(tn, tn + fn),
    }

    return metrics


def build_comparative_metrics_table(
    y_true: pd.Series,
    predictions: dict[str, np.ndarray],
) -> pd.DataFrame:
    """Construye una tabla comparativa con las métricas solicitadas.

    Args:
        y_true: Etiquetas reales del dataset.
        predictions: Diccionario con predicciones K-Folds por modelo.

    Returns:
        DataFrame comparativo ordenado por precisión global descendente.

    Raises:
        TypeError: Si predictions no es un diccionario.
    """
    if not isinstance(predictions, dict) or not predictions:
        raise TypeError("predictions debe ser un diccionario no vacío.")

    rows = [
        calculate_requested_metrics(y_true, y_pred, model_name)
        for model_name, y_pred in predictions.items()
    ]

    metrics_df = pd.DataFrame(rows)
    metrics_df = metrics_df.sort_values("Precisión Global", ascending=False).reset_index(drop=True)
    return metrics_df
```

```python
# Tabla comparativa de métricas solicitadas.

metrics_df = build_comparative_metrics_table(y, kfolds_predictions)

metrics_path = OUTPUT_TABLES / "metricas_kfolds_modelos.csv"
metrics_df.to_csv(metrics_path, index=False)

display_responsive_table(
    metrics_df.round(4),
    title="Tabla comparativa de métricas K-Folds",
    min_width_px=980,
    index=False,
)

print(f"Tabla guardada en: {format_relative_path(metrics_path)}")
```
